```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(1234)
```

```python
# genrate training data
N = 10
# input values for training
x = np.linspace(start=0.0, stop=1.0, num=N)
x = np.expand_dims(x, axis=-1)

# target values for training
y = np.sin(2*np.pi*x) 
# add some random noise
y = y + np.random.normal(loc=0.0, scale=0.1, size=y.shape)


# generate test data, which are used to plot a true target function. 
test_x = np.linspace(start=0.0, stop=1.0, num=100)
test_x = np.expand_dims(test_x, axis=-1)
test_y = np.sin(2*np.pi*test_x)


# the order of polynomial
M = 0
X = [] # transformed inptus for training
for m in range(M+1):
    x_ = np.power(x, m)
    X.append(x_)
X = np.concatenate(X, axis=-1)


test_X = []
for m in range(M+1):
    x_ = np.power(test_x, m)
    test_X.append(x_)
test_X = np.concatenate(test_X, axis=-1)


# write codes to compute correct 'w'.
XX = np.matmul(X.T, X)
iXX = np.linalg.inv(XX)
iXXX = np.matmul(iXX, X.T)
w = np.matmul(iXXX, y)


y_est = np.matmul(test_X, w)

plt.figure()
plt.plot(x, y, 'bo')
plt.plot(test_x, test_y, 'g-')
plt.plot(test_x, y_est, 'r-')
plt.show()
```
